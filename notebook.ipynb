```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
from matplotlib import gridspec 
```

```python
data = pd.read_csv("/datasets/awesome/creditcard.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
fraud = data[data['Class'] == 1]
valid = data[data['Class'] == 0]
outlierFraction = len(fraud)/float(len(valid))
print(outlierFraction)
print("Fraud Cases: {}".format(len(fraud)))
print("Valid Transactions: {}".format(len(valid)))
```

```python
print("Amount details in fradudulent transactions:")
print(fraud.Amount.describe())
```

```python
print("Amount details in valid transactions:")
print(valid.Amount.describe())
```

```python
corrmat = data.corr()
fig = plt.figure(figsize = (12, 9))
sns.heatmap(corrmat, vmax=.8, square=True)
plt.show()
```

```python
X = data.drop(['Class'], axis = 1)
Y = data['Class']
print(X.shape)
print(Y.shape)
xData = X.values
yData = Y.values
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(xData, yData, test_size = 0.2, random_state = 42)
```

```python
pip install  scikit-learn
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(xTrain, yTrain)
yPred = rfc.predict(xTest)
```

```python
from sklearn.metrics import classification_report, accuracy_score  
from sklearn.metrics import precision_score, recall_score 
from sklearn.metrics import f1_score, matthews_corrcoef 
from sklearn.metrics import confusion_matrix 
```

```python
n_outliers = len(fraud)
```

```python
n_errors = 0
for (i, j) in zip(yTest, yPred):
    if i != j:
        n_errors = n_errors + 1

print(n_errors)
print("The model used is random forest classifier")
acc = accuracy_score(yTest, yPred)
print("The accuracy is {}".format(acc))
prec = precision_score(yTest, yPred) 
print("The precision is {}".format(prec)) 
  
rec = recall_score(yTest, yPred) 
print("The recall is {}".format(rec)) 
  
f1 = f1_score(yTest, yPred) 
print("The F1-Score is {}".format(f1)) 
  
MCC = matthews_corrcoef(yTest, yPred) 
print("The Matthews correlation coefficient is{}".format(MCC)) 
```

```python
LABELS = ['Normal', 'Fraud']
conf_matrix = confusion_matrix(yTest, yPred)
plt.figure(figsize = (12, 9))
sns.heatmap(conf_matrix, xticklabels=LABELS, annot = True, yticklabels=LABELS, fmt = "d")
plt.title("Confusion Matrix")
plt.xlabel("True Class")
plt.ylabel("Predicted Class")
plt.show()
```
